```python
import pandas as pd

# Monta Google Drive si es necesario
# from google.colab import drive
# drive.mount('/content/drive')

# Carga el archivo CSV
file_path = '/content/country_vaccinations1.csv'
 # Reemplaza con la ruta real
df = pd.read_csv(file_path)

# Muestra las primeras filas del dataframe
df.head()

# Muestra la información del dataframe
df.info()

# Muestra estadísticas descriptivas de las columnas numéricas
df.describe()

# Muestra estadísticas descriptivas de las columnas categóricas
df.describe(include='object')

# Agrupar por la columna 'vaccines' y sumar la cantidad de vacunas aplicadas
vacunas_por_compania = df.groupby('vaccines')['total_vaccinations'].sum().reset_index()

# Mostrar el resultado
vacunas_por_compania

# Calcular la cantidad total de vacunas aplicadas en todo el mundo
total_vacunas_mundo = df['total_vaccinations'].sum()

# Mostrar el resultado
print("Cantidad total de vacunas aplicadas en todo el mundo:", total_vacunas_mundo)

# Calcular el promedio de vacunas aplicadas por país
promedio_por_pais = df.groupby('country')['total_vaccinations'].mean()

# Mostrar el resultado
print("Promedio de vacunas aplicadas por país:")
print(promedio_por_pais)

# Convertir la columna 'date' a formato datetime64
df['date'] = pd.to_datetime(df['date'])

# Filtrar el dataframe para obtener solo los datos del 29/01/21
vacunas_29_01_21 = df[df['date'] == '2021-01-29']

# Calcular la cantidad total de vacunas aplicadas en todo el mundo ese día
total_vacunas_29_01_21 = vacunas_29_01_21['total_vaccinations'].sum()

# Mostrar el resultado
print("Cantidad de vacunas aplicadas el 29/01/21 en todo el mundo:", total_vacunas_29_01_21)

# Crear un nuevo dataframe conDiferencias
conDiferencias = df.copy()

# Calcular las diferencias entre las columnas daily_vaccinations y daily_vaccinations_raw
conDiferencias['diferencias'] = df['daily_vaccinations'] - df['daily_vaccinations_raw']

# Mostrar el nuevo dataframe
print(conDiferencias.head())

# Obtener la fecha más reciente y la fecha más antigua
fecha_mas_reciente = df['date'].max()
fecha_mas_antigua = df['date'].min()

# Calcular el periodo de tiempo entre la fecha más reciente y la fecha más antigua
periodo_tiempo = fecha_mas_reciente - fecha_mas_antigua

# Mostrar el resultado
print("Fecha más reciente:", fecha_mas_reciente)
print("Fecha más antigua:", fecha_mas_antigua)
print("Periodo de tiempo:", periodo_tiempo)

# Separar la columna 'vaccines' por el carácter ',' y expandir en nuevas columnas
vacunas_por_dia = df['vaccines'].str.split(', ', expand=True)

# Obtener la cantidad de vacunas utilizadas cada día sumando las nuevas columnas
df['canVac'] = vacunas_por_dia.count(axis=1)

# Crear un nuevo dataframe denominado conCantidad con los datos originales y la nueva columna 'canVac'
conCantidad = df.copy()

# Mostrar el nuevo dataframe conCantidad
print(conCantidad)

# Convertir la columna 'date' al formato datetime
df['date'] = pd.to_datetime(df['date'])

# Filtrar los registros antes del 20 de diciembre de 2020
antes20 = df[df['date'] < '2020-12-20']

# Mostrar el nuevo dataframe antes20
print(antes20)

# Filtrar los registros que contienen la vacuna Pfizer
pfizer = df[df['vaccines'].str.contains('Pfizer')]

# Mostrar el nuevo dataframe pfizer
print(pfizer)

# Importar la librería pandas
import pandas as pd

# Crear un objeto ExcelWriter
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    # Almacenar los dataframes en hojas separadas
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)


```
